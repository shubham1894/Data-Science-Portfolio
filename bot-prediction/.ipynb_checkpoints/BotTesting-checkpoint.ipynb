## Handling Missing Value
### Just go serially one by one from left to right

```python
# First fill missing value in secondary Domain
test_df['sec_lvl_domn'] = test_df['sec_lvl_domn'].fillna(test_df['sec_lvl_domn'].mode()[0])
```

### Now moving ahead device type and page_vw_ts.   Droping these two value because device_type has lots of missing value arround 80%.    Also droping the page_vw_ts because timing doesn't matter in this case and   also it is hard to find  mode because timing is in seconds and milliseconds which eventually give us many different values.

```python
test_df.drop(['device_type','page_vw_ts'],axis=1, inplace= True)
```
