#### Target Guided Ordinal Encoding
1. Ordering the label according to the target
2. Replace the labels by the joint probablity of being 1 or 0

```python
ordinal_label1 = df.groupby(['ctry_name'])['target'].mean().sort_values().index
ordinal_label2 = df.groupby(['city'])['target'].mean().sort_values().index
ordinal_label3 = df.groupby(['st'])['target'].mean().sort_values().index
ordinal_label4 = df.groupby(['sec_lvl_domn'])['target'].mean().sort_values().index
ordinal_label5 = df.groupby(['operating_sys'])['target'].mean().sort_values().index
ordinal_label6 = df.groupby(['ip_addr'])['target'].mean().sort_values().index
ordinal_label7 = df.groupby(['user_agent'])['target'].mean().sort_values().index
ordinal_label8 = df.groupby(['page_url'])['target'].mean().sort_values().index
```

```python
ordinal_label11 = {k:i for i,k in enumerate(ordinal_label1,0)}
ordinal_label22 = {k:i for i,k in enumerate(ordinal_label2,0)}
ordinal_label33 = {k:i for i,k in enumerate(ordinal_label3,0)}
ordinal_label44 = {k:i for i,k in enumerate(ordinal_label4,0)}
ordinal_label55 = {k:i for i,k in enumerate(ordinal_label5,0)}
ordinal_label66 = {k:i for i,k in enumerate(ordinal_label6,0)}
ordinal_label77 = {k:i for i,k in enumerate(ordinal_label7,0)}
ordinal_label88 = {k:i for i,k in enumerate(ordinal_label8,0)}
```

```python
df['ctry_name'] = df['ctry_name'].map(ordinal_label11)
df['city'] = df['city'].map(ordinal_label22)
df['st'] = df['st'].map(ordinal_label33)

df['sec_lvl_domn'] = df['sec_lvl_domn'].map(ordinal_label44)
df['operating_sys'] = df['operating_sys'].map(ordinal_label55)
df['ip_addr'] = df['ip_addr'].map(ordinal_label66)
df['user_agent'] = df['user_agent'].map(ordinal_label77)
df['page_url'] = df['page_url'].map(ordinal_label88)

df.head()
```

```python
# Drop unnecessary column 
# Input Split 
X = train_df2.drop(columns=['index','target'], axis=1)
y = train_df2['target']
```

#### Model Training 

```python
# classify column
def classify(model):
    x_train,x_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state= 42)
    model.fit(x_train,y_train)
    print('Accuracy:', model.score(x_test,y_test))
    
    score = cross_val_score(model, X, y, cv= 5)
#     print(score)
    print("CV Score:", np.mean(score))
    
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
classify(model)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
classify(model)
```
